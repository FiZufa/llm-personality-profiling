## Zero shot results

```python
import pandas as pd
import numpy as np
```

```python
result_gpt_16type = pd.read_json("result_gpt_16type.json")
result_gpt_IE = pd.read_json("result_gpt_IE.json")
result_gpt_NS_1 = pd.read_json("result_gpt_NS_170.json")
result_gpt_NS_2 = pd.read_json("result_gpt_NS_171.json")
result_gpt_NS = pd.concat([result_gpt_NS_1, result_gpt_NS_2], ignore_index=True)
result_gpt_FT = pd.read_json("result_gpt_FT.json")
result_gpt_JP = pd.read_json("result_gpt_JP.json")

result_ds_16type_1 = pd.read_json("result_deepseek_16type_500.json")
result_ds_16type_2 = pd.read_json("result_deepseek_16type_501.json")
result_ds_16type = pd.concat([result_ds_16type_1, result_ds_16type_2], ignore_index=True)
result_ds_IE = pd.read_json("result_deepseek_IE.json")
result_ds_NS = pd.read_json("result_deepseek_NS.json")
result_ds_FT = pd.read_json("result_deepseek_FT.json")
result_ds_JP = pd.read_json("result_deepseek_JP.json")
```

```python
result_ds_16type = result_ds_16type.rename(columns={'content': 'response'})
```

```python
def count_binary_acc(df, idx):
    response_trait = df['response']
    actual_trait = df['mbti_actual'].str[idx]

    # count how many data response_trait==actual_trait
    correct_matches = (response_trait == actual_trait)
    n_correct = correct_matches.sum()

    acc = n_correct / len(df)

    print(f"# correct = {n_correct}/1062, accuracy = {acc}")
```

```python
def count_16type_acc(df):

    correct_matches = (df['response'] == df['mbti_actual'])
    n_correct = correct_matches.sum()

    acc = n_correct / len(df)

    print(f"# correct = {n_correct}/1062, accuracy = {acc}")
```

### GPT-4o

```python
print('16 Types')
count_16type_acc(result_gpt_16type)
print('--------------------------')
print('I vs E')
count_binary_acc(result_gpt_IE, 0)
print('--------------------------')
print('N vs S')
count_binary_acc(result_gpt_NS, 1)
print('--------------------------')
print('F vs T')
count_binary_acc(result_gpt_FT, 2)
print('--------------------------')
print('J vs P')
count_binary_acc(result_gpt_JP, 3)
print('--------------------------')
```

### Deepseek-R1

```python
print('16 Types')
count_16type_acc(result_ds_16type)
print('--------------------------')
print('I vs E')
count_binary_acc(result_ds_IE, 0)
print('--------------------------')
print('N vs S')
count_binary_acc(result_ds_NS, 1)
print('--------------------------')
print('F vs T')
count_binary_acc(result_ds_FT, 2)
print('--------------------------')
print('J vs P')
count_binary_acc(result_ds_JP, 3)
print('--------------------------')
```

### LlaMa-3.1-8b

```python
result_llama_df = pd.read_csv('result_llama_16type_all.csv')
```

```python
result_llama_df
```

```python
def clean_data(df, col_name):

    # take only the first character of its columns
    df[col_name] = df[col_name].astype(str).str[0]
    return df
```

```python
result_llama_df = clean_data(result_llama_df, 'IE_pred')
```

```python
result_llama_df = clean_data(result_llama_df, 'NS_pred')
result_llama_df = clean_data(result_llama_df, 'FT_pred')
result_llama_df = clean_data(result_llama_df, 'JP_pred')
```

```python
result_llama_df
```

```python
def count_llama_binary_acc(df, col_name, idx):
    correct_matches = (df['type'].str[idx] == df[col_name])
    n_correct = correct_matches.sum()

    acc = n_correct / len(df)
    print(f"# correct = {n_correct}/1062, accuracy = {acc}")

def count_llama_16type_acc(df):
    correct_matches = (df['type'] == df['mbti_pred'])
    n_correct = correct_matches.sum()

    acc = n_correct / len(df)
    print(f"# correct = {n_correct}/1062, accuracy = {acc}")
```

```python
print('16 Types')
count_llama_16type_acc(result_llama_df)
print('--------------------------')
print('I vs E')
count_llama_binary_acc(result_llama_df, 'IE_pred', 0)
print('--------------------------')
print('N vs S')
count_llama_binary_acc(result_llama_df, 'NS_pred', 1)
print('--------------------------')
print('F vs T')
count_llama_binary_acc(result_llama_df, 'FT_pred', 2)
print('--------------------------')
print('J vs P')
count_llama_binary_acc(result_llama_df, 'JP_pred', 3)
print('--------------------------')
```
